# Verification of DBScan clustering of Precipitation Events

## Objective: 
The DBSCAN algorithm was used to group observations from TRMM (from the swath paths) into 'rain events' based on the location and time of rain observations. There hyperparameters were used in this process: the minimum number of observations needed to be considered and 'event', the distance threshold underwhich observations are clustered together, and the conversion of time into a equivalent spatial distance when calculating the distance between observations.

Here, the objective is to evaluate whether the DBSCAN algorithm is producing sensical results. A qualititave approach is taken by comparing snapshots of the clustered output to other satellite products at similar points in time. Both output from the 3-Hour Real Time TRMM Multi-satellite Precipitation Analysis, a gridded data product that includes TRMM observations as well as those from other satellites (https://pmm.nasa.gov/sites/default/files/document_files/3B4XRT_doc_V7_180426.pdf) and the MODIS satellite (access from here: https://ladsweb.modaps.eosdis.nasa.gov/search/) were used.

### Loading in the Data
We start by importing modules, as well as downloading the dbscan output from S3 and importing it into the notebook to work with.

```python
#import the different packages used throughout
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap, cm
import math
import datetime
import boto3
from os.path import expanduser
import os
import json
import glob
from scipy.stats import skew
import cartopy.crs as ccrs


%matplotlib inline

def load_Data():
    #download the compiled dbscan output and return the xarray data
    home = expanduser("~")

    with open(os.path.join(home,'creds.json')) as creds_file:
        creds_data = json.load(creds_file)

    #Access from S3
    s3 = boto3.resource('s3',aws_access_key_id=creds_data['key_id'],
             aws_secret_access_key=creds_data['key_access'],region_name='us-west-2')
    bucket = s3.Bucket('himatdata')
    home = os.getcwd()

    #download the file (note this is a big, 10+ GB file)
    bucket.download_file('Trmm/EPO/Cluster_results/K_DB_Clustered_Data.nc4',os.path.join(os.path.join(home,'S3_downloads/K_DB_Clustered_Data.nc4')))
    #open the local copy
    F = xr.open_dataset('S3_downloads/K_DB_Clustered_Data.nc4')

    return F


F = load_Data()

#data contained in the dataset
print(F.data_vars)
print(F.coords)
```

### Exploring a Specific Day
We looked at the output from January 1, 2009, to get a sense of what observations were being clustered together. First clusters with observations on that day are extracted from the dataset and plotted. We also look at when those observations were taken to understand both the spatial component of the clustering as well as the temporal component.

```python
#helper functions for the analysis

def extract_day(year,month,day,F):
    #function that extracts all clusters in F with at least one observations on a given day
    
    #get the indices of the time variabile for the date of interest
    ind = (F.Time.dt.year==year) & (F.Time.dt.month==month) & (F.Time.dt.day==day)
    #get the unique cluster numbers that correspond to those times
    db_ind = np.unique(F.DBLabel[ind])

    #create empty matrices to hold the subset of data of interes
    subset_dblabels = []
    subset_lat = []
    subset_lon = []
    subset_time = []

    #for each cluster that falls on that day, extract some data and append it to the array
    for i in range(len(db_ind)):
        ind = F.DBLabel==db_ind[i]
        subset_dblabels = np.append(subset_dblabels,F.DBLabel[ind])
        subset_lat = np.append(subset_lat,F.Latitude[ind])
        subset_lon = np.append(subset_lon,F.Longitude[ind])
        subset_time = np.append(subset_time,np.array(F.Time[ind]-np.datetime64('2009-01-01'),'float')/3.6e12)
        #time is converted to hours
    
    return subset_dblabels, subset_lat, subset_lon, subset_time


def plot_clusters_subset(subset_dblabels,subset_lon,subset_lat,subset_time, view='merc'):
    #function that creates a figure with two subplots - one showing observations colored by cluster and the 
    #other showing observations colored by time since midnight Jan 1 2009 the observations were made in hours
    
    #get the cluster numbers
    uniqLabels = np.unique(subset_dblabels)
        
    #region of plot
    bbox = [-40,40,150,260]
    
    #spacing of parallels and meridians in plot
    parallels = np.arange(-60.,81,10.)
    verts = np.arange(0.,360,25.)
    
    fig = plt.figure(figsize=(20,9))
    plt.subplot(1,2,1)
    
    #if the close up view for MODIS comparision, change some values
    if view == 'cass':
        bbox = [0,22,-176,-149]
        parallels = np.arange(0.,20,5.)
        verts = np.arange(-175.,-150,5.)
        # Define the projection, scale, the corners of the map, and the resolution.        
        m = Basemap(projection='cass',llcrnrlat=bbox[0],urcrnrlat=bbox[1],\
            llcrnrlon=bbox[2],urcrnrlon=bbox[3],lat_ts=10,resolution='i',
            lon_0=(bbox[2]+bbox[3])/2,lat_0=(bbox[0]+bbox[1])/2)
    else:
        # Define the projection, scale, the corners of the map, and the resolution.
        m = Basemap(projection=view,llcrnrlat=bbox[0],urcrnrlat=bbox[1],\
                    llcrnrlon=bbox[2],urcrnrlon=bbox[3],lat_ts=10,resolution='i')
    cmap = plt.get_cmap('tab20')

    # Draw coastlines and put in parallels and meridians
    m.drawcoastlines()
    m.drawparallels(parallels,labels=[False,True,True,False]);
    m.drawmeridians(verts,labels=[False,True,True,False]);

    #for each cluster, plot the observations in it as a scatter plot
    for ii in range(len(uniqLabels)):
        color_interp = np.interp(ii,[0,len(uniqLabels)],[0,20])
        m.scatter(np.array(subset_lon[subset_dblabels==uniqLabels[ii]]), 
                  np.array(subset_lat[subset_dblabels==uniqLabels[ii]]), 
                  latlon=True,marker='o',color=cmap(ii),s=5)

    plt.subplot(1,2,2)
    if view == 'cass':
        # Define the projection, scale, the corners of the map, and the resolution.        
        m = Basemap(projection='cass',llcrnrlat=bbox[0],urcrnrlat=bbox[1],\
            llcrnrlon=bbox[2],urcrnrlon=bbox[3],lat_ts=10,resolution='i',
            lon_0=(bbox[2]+bbox[3])/2,lat_0=(bbox[0]+bbox[1])/2)
    else:
        # Define the projection, scale, the corners of the map, and the resolution.
        m = Basemap(projection=view,llcrnrlat=bbox[0],urcrnrlat=bbox[1],\
                    llcrnrlon=bbox[2],urcrnrlon=bbox[3],lat_ts=10,resolution='i')
        
    cmap = plt.get_cmap('tab20')

    # Draw coastlines and put in parallels and meridians
    m.drawcoastlines()
    m.drawparallels(parallels,labels=[False,True,True,False]);
    m.drawmeridians(verts,labels=[False,True,True,False]);

    im = m.scatter(np.array(subset_lon),np.array(subset_lat), latlon=True,marker='o',
                   c=subset_time,s=5)

    #add a colorbar
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(im, cax=cbar_ax, label='Time since Jan 1, 2009 [hours]')
```

```python
subset_dblabels, subset_lat, subset_lon, subset_time = extract_day(2009,1,1,F)
plot_clusters_subset(subset_dblabels,subset_lon,subset_lat,subset_time)
```

Just visually, some funky things are going on here. Most notably, DBSCAN is grouping observations in a swath with another swath basically whenever there is precipitation that somewhat overlaps spatially, even when they're off by several hours (see orange cluster on left as a good example of this, which spans over more than 10 hours).

### Comparision with A Gridded TRMM Product
How does this visually compare to observations from the 3-Hour Real Time TRMM Multi-satellite Precipitation Analysis? Taking all the files that fall on January 1, 2009 from that dataset, I compiled a net precipitation for that day by summing them together. Note that I threw out data that had a rainrate less than 2 mm/hr, as those low rainrates were not included in the DBSCAN clustering. 

```python
#helper function for the analysis
def plot_TRMML3():
    #zeros matrix to create the running total from the day
    rainrate = np.zeros((1440,400))
    
    #read through all files in the folder where I locally downloaded the L3 output for January 1, 2009
    for file in glob.glob('TRMM L3/*.nc4'):
        trmm_gridded = xr.open_dataset(file)
        toadd = np.array(trmm_gridded.precipitation)
        #only look at rainrates above 2 mm/hr to be consistent with the DBSCAN cutoff
        toadd[toadd<2]=0
        rainrate += toadd
        trmm_gridded.precipitation.values = rainrate

    #plot the results and include coastlines    
    fig = plt.figure(figsize=(15,9))
    ax = plt.axes(projection=ccrs.PlateCarree(180))

    p = trmm_gridded.precipitation.where(trmm_gridded.precipitation>0).plot(ax=ax,x='nlon',y='nlat',transform=ccrs.PlateCarree())
    ax.set_extent((150, 260, -40, 40))
    ax.coastlines();
```

```python
plot_TRMML3()
#note that the colorbar label is not in the right units, as I am summing all the rainrates over that day
```

This doesn't seem to match very closely with the results of DBSCAN above. Obviously the swath nature of TRMM data is going to play into that but I'm also a bit bamboozled that I am picking up rain in places the other dataset doesn't indicate rain occurring on that day. Visually the raining groupings here do not correspond with the ones above. There are a number of things that play into this that could result in visual trickery, so it is important to keep them in mind:
1. Comparing apples to oranges. These are not exact comparisions given the filtering out of low rain rates that occurs in both as well as the processing that happens in the gridded product
2. The time of rain events is completely obscured in the gridded product as plotted here. What appear as visual clusters may not be disparate in time

### Meaning from Chaos?
This comparison suggests that DBSCAN (with the current hyperparameters) is grouping too much together into individual clusters. So how to fix it? Obviously it's possible to rerun DBSCAN with different hyperparameters, and there are a number of adjustments that could ammeloriate these issues:
1. Decreasing the distance allowed between points is the biggest lever
2. Fiddle with the scaling from time to km that is used when calculating the distance metric to penalize time disparities more

The question becomes how to better define these parameters? Trial and error is possible but there should be something more robust than that. Also, how do you know when you've got it 'right'? An important piece of the later question is the larger downside associated with clustering things that ought not be clusters as opposed to the downside of breaking up a cluster into too many smaller clusters. Grouping things that aren't really a rain event will skew the mean profiles of latent heat, but breaking up what should be one event into multiples should still give you similar mean profiles (just skewing the statistics of frequency of events).
